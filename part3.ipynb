```python
import pandas as pd
import random as rand
from faker import Faker
```

```python
data = {
    'StudentFName': [Faker().name() for i in range(30)],
    'StudentLName': [Faker().last_name() for i in range(30)],
    'age': [rand.randint(13, 15) for i in range(30)],
    'grades': [rand.randint(40, 100) for i in range(30)]
}
```

```python
data
```

```python
df = pd.DataFrame(data)
```
