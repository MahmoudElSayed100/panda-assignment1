```python
builtAfter2000 = df[df['yr_built']>2000]
pbb = builtAfter2000[['price']]
toph = pbb.head(10)
print(toph)
avgPbb = toph['price'].mean()
print(avgPbb)
```

```python
withView = df[df.view == 1]
noView = df[df.view == 0]
prcWV= withView['price'].mean()
prcNV= noView['price'].mean()
print("With view ", prcWV)
print("Without view", prcNV)
```

```python
withW = df[df.waterfront == 1]
noW = df[df.waterfront == 0]
prcWV= withW['price'].mean()
prcNV= noW['price'].mean()
print("With waterfront ", prcWV)
print("Without waterfront", prcNV)
```
